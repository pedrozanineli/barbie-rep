```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from lmfit import Model
PALETTE = ['#1b6872', '#df208a', '#efb475', '#aec5fb', '#D5A6E6', '#A4D4B4', '#FFD166', '#FF6F61']
```

```python
from matplotlib import font_manager
font_path = 'Ruda-Regular.ttf'
font_prop = font_manager.FontProperties(fname=font_path)
```

# BaCBM2

```python
df_trp = pd.read_excel('Kd_trp/trp quench.xlsx', header = 0)

PS = df_trp['[PS] ng/mL'][0:11]
# Norm_fluor = df_trp['CBM normalized fluorescence'][0:11]
Norm_fluor = df_trp['CBM1'][0:11]

std_fluor = df_trp['Std_fluorescence'][0:11]

# Dados fornecidos
x = np.array(PS)
y = 100*np.array(Norm_fluor)
y_std = 100*np.array(std_fluor)

# Definição da função não-linear
def model_func(x, a, b):
    return a * x / (b + x)

# Criação do modelo usando lmfit
model = Model(model_func)

# Definir valores iniciais e ajustar
params = model.make_params(a=100, b=100)  # Chute inicial para a e b
result = model.fit(y, params, x=x)

# Parâmetros ajustados
a = result.params['a'].value
b = result.params['b'].value
# R2 = result.rsquared

# Geração de dados ajustados
x_fit = np.linspace(0, max(x), 100)
y_fit = result.eval(x=x_fit)

# Create the plot again for saving and displaying
plt.figure(figsize=(5, 3), dpi=100)
plt.errorbar(x, y, yerr=y_std, fmt='none', ecolor=PALETTE[1])
plt.scatter(x, y, color=PALETTE[1], s=20)
# plt.plot(x_fit, y_fit, color=PALETTE[0], linewidth=2, linestyle='--', label = f'R$^2$ = {round(R2, 2)}\nK$_d$ = {round(b, 2)} (mg L$^-$$^1$)')
plt.plot(x_fit, y_fit, color=PALETTE[0], linewidth=2, linestyle='--', label = f'K$_d$ = {round(b, 2)} (mg L$^-$$^1$)')

# Classic styling: grid and labels without background color
plt.grid(True, linestyle='--', linewidth=0.3, color='gray', alpha=0.8)
#plt.title('Classic 2D Graph', fontsize=16, pad=5, fontproperties=font_prop)
plt.xlabel('[PS] (ng mL$^{-1})$', fontsize=16)
plt.ylabel('Normalized Emission (%)', fontsize=16)
plt.legend(fontsize=12, loc=4, framealpha=0.5)

# Hide the right and top spines
plt.gca().spines[['right', 'top']].set_visible(False)

# Save the plot as an SVG file
#plt.savefig('fit_langmuir.png', format='png', bbox_inches='tight', transparent=True, dpi=600)

# Show the plot
plt.show()
```
