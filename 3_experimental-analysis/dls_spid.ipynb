```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import font_manager

font_path = '../Ruda/Ruda-VariableFont_wght.ttf'
font_manager.fontManager.addfont(font_path)
prop = font_manager.FontProperties(fname=font_path)

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = prop.get_name()
```

```python
df = pd.read_excel('DLS/iGEM.xlsx')
```

```python
def proc(name):

    i_start,i_end = list(df['Sample Name']).index(name),list(df['Sample Name']).index(name[:-1]+'3')
    df_new = df.loc[i_start:i_end]
    df_new = df_new.loc[:,'Intensities[1] (Percent)':'Intensities[300] (Percent)']
    df_new_mean = df_new.mean()

    return list(df_new_mean)
```

```python
list(df['Sample Name'])
```

```python
tubo1_medida1 = 'TUBO1_MEDIDA1_17092024_SPIDROINA 1'
tubo1_medida3 = 'TUBO1_MEDIDA2_17092024_SPIDROINA 1'

tubo2_medida1 = 'TUBO2_MEDIDA1_17092024_SPIDROINA 1'
tubo2_medida3 = 'TUBO2_MEDIDA2_17092024_SPIDROINA 1'

tubo3_medida1 = 'TUBO3_MEDIDA1_17092024_SPIDROINA 1'
tubo3_medida3 = 'TUBO3_MEDIDA2_17092024_SPIDROIN 1'

tubo4_medida1 = 'TUBO4_MEDIDA1_17092024_SPIDROINA 1'
tubo4_medida3 = 'TUBO4_MEDIDA2_17092024_SPIDROINA 1'

t1_m1 = proc(tubo1_medida1)
t1_m3 = proc(tubo1_medida3)

t2_m1 = proc(tubo2_medida1)
t2_m3 = proc(tubo2_medida3)

t3_m1 = proc(tubo3_medida1)
t3_m3 = proc(tubo3_medida3)

t4_m1 = proc(tubo4_medida1)
t4_m3 = proc(tubo4_medida3)
```

```python
# size = list(df.iloc[0,307:606].values)
size = range(len(t1_m1))
```

```python
size = np.linspace(1,1000,300)
```

```python
plt.plot(size,np.array(t1_m1)/10,label='Before Heating',alpha=1,color='#477081')
plt.plot(size,np.array(t1_m3)/10,label='After Heating',alpha=1,color='#EFB475')
plt.title('Tube 1',fontsize=16)
plt.xlabel('Diameter (nm)',fontsize=14)
plt.ylabel('Intensity (%)',fontsize=14)
plt.xlim(1e1,1e3)
plt.legend(fontsize=14,loc='upper left')
plt.xscale('log')
plt.grid()
plt.savefig('results/t1.png',transparent=True,dpi=800)
plt.show()

plt.plot(size,np.array(t2_m1)/10,label='Before Heating',alpha=1,color='#477081')
plt.plot(size,np.array(t2_m3)/10,label='After Heating',alpha=1,color='#EFB475')
plt.title('Tube 2',fontsize=16)
plt.xlabel('Diameter (nm)',fontsize=14)
plt.ylabel('Intensity (%)',fontsize=14)
plt.xlim(1e1,1e3)
plt.legend(fontsize=14,loc='upper left')
plt.xscale('log')
plt.grid()
plt.savefig('results/t2.png',transparent=True,dpi=800)
plt.show()

plt.plot(size,np.array(t3_m1)/10,label='Before Heating',alpha=1,color='#477081')
plt.plot(size,np.array(t3_m3)/10,label='After Heating',alpha=1,color='#EFB475')
plt.title('Tube 3',fontsize=16)
plt.xlabel('Diameter (nm)',fontsize=14)
plt.ylabel('Intensity (%)',fontsize=14)
plt.xlim(1e1,1e3)
plt.legend(fontsize=14,loc='upper left')
plt.xscale('log')
plt.grid()
plt.savefig('results/t3.png',transparent=True,dpi=800)
plt.show()

plt.plot(size,np.array(t4_m1)/10,label='Before Heating',alpha=1,color='#477081')
plt.plot(size,np.array(t4_m3)/10,label='After Heating',alpha=1,color='#EFB475')
plt.title('Tube 4',fontsize=16)
plt.xlabel('Diameter (nm)',fontsize=14)
plt.ylabel('Intensity (%)',fontsize=14)
plt.xscale('log')
plt.xlim(1e1,1e3)
plt.legend(fontsize=14,loc='upper left')
plt.grid()
plt.savefig('results/t3.png',transparent=True,dpi=800)
plt.show()
```
