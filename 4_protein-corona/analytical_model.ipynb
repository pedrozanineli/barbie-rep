```python

kB = 1.380649e-23
T = 300
beta = kB*T

v_avg_prot = 5 # descubra
v_avg_part = 5 # descubra
v_avg_sis = 5

protein_radius = 1.25
mp_radius = 50
n_max = 40000 # qtd proteinas na area superficial
# n = # particulas que ja se aderiram
epsilon = -0.44 # energia de interacao CBM-PS

densidade_poliestireno = 15
volume_part = 4/3*np.pi*mp_radius**3
m_part = volume_part * densidade_poliestireno

m_prot = 7.45e-23

dn_dt = []

m_total = m_part

prot_conc = 0.007

n = 100

for n in range(100):

    alpha = (m_part*v_avg_part + m_prot*v_avg_prot)/(m_total+m_prot) # momento medio relacionado a velocidade media
    result = prot_conc * v_avg_sis * np.pi * (protein_radius + mp_radius)**2 * (n_max-n)/n_max * 1/(1+np.exp(beta*(epsilon-alpha)))
    dn_dt.append(result)
    m_total = m_part+m_prot
```

```python
plt.plot(range(n),dn_dt[:-1])
plt.show()
```
