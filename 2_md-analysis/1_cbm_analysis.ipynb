```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import font_manager

font_path = '../Ruda/Ruda-VariableFont_wght.ttf'
font_manager.fontManager.addfont(font_path)
prop = font_manager.FontProperties(fname=font_path)

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = prop.get_name()
```

#### Distance CBM x PS

```python
cbm = pd.read_csv('data/b1_x_cbm/distances_cbm-ps.txt',sep='\t')
b1 = pd.read_csv('data/b1_x_cbm/distances_b1-ps.txt',sep='\t')
```

```python
plt.plot(cbm['Frame'],cbm['Min_Distance'],label='BaCBM2',color='#477081',alpha=0.9,marker='s',markersize=2)
plt.plot(b1['Frame'],b1['Min_Distance'],label='BARBIE1',color='#E0218A',alpha=0.9,marker='s',markersize=2)
plt.title('Distances Protein - Polystyrene',fontsize=16,loc='left')
plt.xlim(0,500)
plt.ylim(0,45)
plt.xlabel('Frame',fontsize=14),plt.ylabel('Distance ()',fontsize=14)

plt.grid(),plt.legend(fontsize=12),plt.tight_layout()
plt.savefig('results/min_dis.png',dpi=500,transparent=True)
plt.show()
```

```python
print('Distncias mnimas mdias (A):')
print(round(np.mean(b1['Min_Distance'][21:]),2))
print(round(np.mean(cbm['Min_Distance'][21:]),2))
```

importante falar que medimos a distancia porque por ser uma simulacao classica, nao existira ligacao de fato

#### RMSD PS

```python
rmsd_ps_b1 = pd.read_csv('data/b1_x_cbm/ps_rmsd-barbie1_ps.dat',sep='\t',names=['ref_mol','ref_frame','mol','frame','rmsd'],skiprows=1)
rmsd_ps_cbm = pd.read_csv('data/b1_x_cbm/ps_rmsd-cbm_ps.dat',sep='\t',names=['ref_mol','ref_frame','mol','frame','rmsd'],skiprows=1)
```

```python
plt.plot(rmsd_ps_cbm['frame'].iloc[:681],rmsd_ps_cbm['rmsd'].iloc[:681],marker='.',label='BaCBM2',color='#477081')
plt.plot(rmsd_ps_b1['frame'].iloc[:681],rmsd_ps_b1['rmsd'].iloc[:681],marker='.',label='Barbie1',color='#E0218A')
plt.title('RMSD Polystyrene',fontsize=16,loc='left')

plt.xlim(0,500)
plt.ylim(0,45)

plt.xlabel('Frame',fontsize=14),plt.ylabel('RMSD',fontsize=14)
plt.grid(),plt.legend(fontsize=12),plt.tight_layout()
plt.savefig('results/ps_RMSD.png',dpi=500,transparent=True)
plt.show()
```
