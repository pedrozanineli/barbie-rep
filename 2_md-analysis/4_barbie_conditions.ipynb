```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import font_manager

font_path = '../Ruda/Ruda-VariableFont_wght.ttf'
font_manager.fontManager.addfont(font_path)
prop = font_manager.FontProperties(fname=font_path)

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = prop.get_name()
```

#### RMSD PS pH

```python
control = pd.read_csv('data/conditions/ps_rmsd_control.dat',sep='\t')
```

```python
ps_pH1 = pd.read_csv('data/conditions/pH/ps_rmsd_pH1.dat',sep='\t')
ps_pH5 = pd.read_csv('data/conditions/pH/ps_rmsd_pH5.dat',sep='\t')
ps_pH9 = pd.read_csv('data/conditions/pH/ps_rmsd_pH9.dat',sep='\t')
```

```python
# PBC Outliers

ps_pH1 = ps_pH1[ps_pH1['   rmsd'] < 6]
ps_pH5 = ps_pH5[ps_pH5['   rmsd'] < 6]
ps_pH9 = ps_pH9[ps_pH9['   rmsd'] < 6]
```

```python
plt.scatter(control['frame'].iloc[:1001],control['   rmsd'].iloc[:1001],label='Control',alpha=0.4,color='#E0218A',s=6)#,marker='s',markersize=2)

plt.scatter(ps_pH1['frame'].iloc[:1001],ps_pH1['   rmsd'].iloc[:1001],label='pH=1',alpha=0.4,color='#EFB475',s=6)#,marker='s',markersize=2)
plt.scatter(ps_pH5['frame'].iloc[:1001],ps_pH5['   rmsd'].iloc[:1001],label='pH=5',alpha=0.4,color='#477081',s=6)#,marker='.',markersize=2)
plt.scatter(ps_pH9['frame'].iloc[:1001],ps_pH9['   rmsd'].iloc[:1001],label='pH=9',alpha=0.4,color='#AEC5FB',s=6)#,marker='s',markersize=2)

plt.grid(),plt.xlabel('Time (ns)',fontsize=14),plt.ylabel('PS RMSD (Å)',fontsize=14),plt.legend()
plt.xlim(-1,1000),plt.ylim(0,7),plt.tight_layout()

plt.savefig('results/RMSD_PS_pH.png',dpi=500)

plt.show()
```
