```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import font_manager

font_path = '../Ruda/Ruda-VariableFont_wght.ttf'
font_manager.fontManager.addfont(font_path)
prop = font_manager.FontProperties(fname=font_path)

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = prop.get_name()
```

```python
df = pd.read_excel('data/KDs.xlsx')
```

```python
df
```

```python
plt.figure(figsize=(8,5))
x = np.arange(len(df['Plastic'].iloc[:6]))
bar_width = 0.35

plt.bar(x, df['BARBIE1'].iloc[:6], width=bar_width, label='BARBIE1', color='#E0218A')
plt.bar(x + 1.1*bar_width, df['BaCBM2'].iloc[:6], width=bar_width, label='BaCBM2', color='#477081')

plt.xticks(x + bar_width / 2, df['Plastic'].iloc[:6],fontsize=18)
plt.yticks(fontsize=14)

plt.xlabel('Plastic',fontsize=18),plt.ylabel('Dissociation Constant (kcal/mol)',fontsize=16)
plt.legend(fontsize=18)

ax = plt.gca()

ax.spines['top'].set_position('zero')
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')

plt.xticks(fontsize=14)

plt.tight_layout()
plt.savefig('results/tcc_kd_plastics.svg',dpi=700,transparent=False)

plt.show()
```
