### Barbie Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Energetic Equilibrium

```python
df = pd.read_csv('barbie-solution-builder/gromacs/potential_energy.xvg',header=23,names=['Time (ps)','Energy (kJ/mol)'],sep='  ')
```

```python
plt.figure(figsize=(8,6))

plt.plot(df['Time (ps)'],df['Energy (kJ/mol)'])

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.xlabel('Time (ps)',fontsize=14)
plt.ylabel('Energy (kJ/mol)',fontsize=14)

plt.ylim(min(df['Energy (kJ/mol)']),max(df['Energy (kJ/mol)'])+100)

plt.grid()
plt.tight_layout()

plt.savefig('results/barbie-equilibrium.png',dpi=300)

plt.show()
```
